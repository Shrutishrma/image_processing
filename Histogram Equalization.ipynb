# Histogram Equilization
# Q1 : Write python code to output the following image and Histogram

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a high contrast image (black and white square in the center)
high_contrast_img = np.zeros((256, 256), dtype=np.uint8)
cv2.rectangle(high_contrast_img, (64, 64), (192, 192), 255, -1)

# Create a low contrast image (gray circle on gray background)
low_contrast_img = np.full((256, 256), 120, dtype=np.uint8)  # Gray background
cv2.circle(low_contrast_img, (128, 128), 60, 140, -1)  # Slightly lighter gray circle

# Plot the images and their histograms
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Display high contrast image
axs[0, 0].imshow(high_contrast_img, cmap='gray')
axs[0, 0].set_title("High Contrast Image")
axs[0, 0].axis('off')

# Display low contrast image
axs[1, 0].imshow(low_contrast_img, cmap='gray')
axs[1, 0].set_title("Low Contrast Image")
axs[1, 0].axis('off')

# Plot histogram of high contrast image
axs[0, 1].hist(high_contrast_img.ravel(), bins=256, range=[0, 256])
axs[0, 1].set_title("Histogram of High Contrast Image")

# Plot histogram of low contrast image
axs[1, 1].hist(low_contrast_img.ravel(), bins=256, range=[0, 256])
axs[1, 1].set_title("Histogram of Low Contrast Image")

plt.tight_layout()
plt.show()
```

# Q2: Demonstrate color models In Histogram equilization

```python
import cv2
import matplotlib.pyplot as plt

# Read the original color image
image = cv2.imread("C:\\Users\\kobik\\Downloads\\th (55).jpg")

# Convert the image from BGR to YCrCb color space
ycrcb_img = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

# Histogram equalization on the Y channel (luminance)
ycrcb_img[:, :, 0] = cv2.equalizeHist(ycrcb_img[:, :, 0])

# Convert back to BGR after equalization
equalized_ycrcb_img = cv2.cvtColor(ycrcb_img, cv2.COLOR_YCrCb2BGR)

# Convert the image from BGR to HSV color space
hsv_img = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Histogram equalization on the V channel (brightness)
hsv_img[:, :, 2] = cv2.equalizeHist(hsv_img[:, :, 2])-

# Convert back to BGR after equalization
equalized_hsv_img = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2BGR)

# Plot the original image and equalized images
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# YCrCb Equalized Image
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(equalized_ycrcb_img, cv2.COLOR_BGR2RGB))
plt.title('Equalized in YCrCb Color Model')
plt.axis('off')

# HSV Equalized Image
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(equalized_hsv_img, cv2.COLOR_BGR2RGB))
plt.title('Equalized in HSV Color Model')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Q3: Demonstrate the concept of contrast-limited adaptive histogram equalization (CLAHE) to improve image quality.
