# 3. Demonstrate with suitable applications where adaptive and Otsu's method of thresholding is useful.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the image in color
image_color = cv2.imread("C:\\Users\\kobik\\Downloads\\th (61).jpg", cv2.IMREAD_COLOR)

# Convert to RGB (OpenCV loads images in BGR by default, so we convert it to RGB for correct color display in matplotlib)
image_color_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

# Convert the image to grayscale for thresholding
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)


# Adaptive Mean Thresholding
adaptive_mean_thresh = cv2.adaptiveThreshold(image_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                             cv2.THRESH_BINARY, 11, 2)

# Adaptive Gaussian Thresholding
adaptive_gaussian_thresh = cv2.adaptiveThreshold(image_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                                 cv2.THRESH_BINARY, 11, 2)

# Step 3: Apply thresholding masks to the original color image
def apply_threshold_to_color(color_image, threshold_mask):
    # Create a mask where the thresholded areas are white (255) and apply it to each channel
    color_thresh = cv2.bitwise_and(color_image, color_image, mask=threshold_mask)
    return color_thresh

adaptive_mean_thresh_color = apply_threshold_to_color(image_color_rgb, adaptive_mean_thresh)
adaptive_gaussian_thresh_color = apply_threshold_to_color(image_color_rgb, adaptive_gaussian_thresh)

# Step 4: Display the original and thresholded color images
plt.figure(figsize=(12, 8))

# Original Color Image
plt.subplot(1, 3, 1)
plt.title('Original Image (Color)')
plt.imshow(image_color_rgb)
plt.axis('off')


# Adaptive Mean Thresholding (Color)
plt.subplot(1, 3, 2)
plt.title('Adaptive Mean Thresholding (Color)')
plt.imshow(adaptive_mean_thresh_color)
plt.axis('off')

# Adaptive Gaussian Thresholding (Color)
plt.subplot(1, 3, 3)
plt.title('Adaptive Gaussian Thresholding (Color)')
plt.imshow(adaptive_gaussian_thresh_color)
plt.axis('off')

plt.tight_layout()
plt.show()
```
