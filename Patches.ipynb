```python
import matplotlib.pyplot as plt
from matplotlib import image

# Load the image
data = image.imread(r"C:\Users\shrut\Downloads\DSC02908.JPG")

# Plot a single line
x1 = [2540, 1550]
y1 = [900, 900]
plt.plot(x1, y1, label='Single Line', color='blue', linewidth=3)

# Plot intersecting lines
x2 = [2500, 2200]
y2 = [2540, 1550]
plt.plot(x2, y2, label='Intersecting Line 1', color='red', linestyle='--', linewidth=2)

x3 = [1650,2250]
y3 = [1560, 250]
plt.plot(x3, y3, label='Intersecting Line 2', color='green', linestyle='-.', linewidth=2)

# Add legend and display
plt.legend()
plt.imshow(data)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import image
import matplotlib.patches as patches

# Load the image using imread (this directly loads it as a NumPy array)
data = image.imread(r"C:\Users\shrut\Downloads\DSC02908.JPG")

# Create figure and axis for plotting
fig, ax = plt.subplots(1)

# Display the image
ax.imshow(data)

# Create a red rectangle and a blue circle
rect = patches.Rectangle((250, 210), 120, 100, linewidth=4, edgecolor='r', facecolor='none')
circle = patches.Circle((1510, 1280), radius=50, linewidth=4, edgecolor='b', facecolor='none')

# Add shapes to the axes
ax.add_patch(rect)
ax.add_patch(circle)

# Show the final image with shapes
plt.show()
```
