```python
import cv2
import matplotlib.pyplot as plt
# Step 1: Read the image and convert to grayscale

image = cv2.imread('number plate.webp') # Replace with your image path
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Step 2 : Apply thresholding
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
# Step 3: Find contours
contours, _ = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# Step 4: Draw contours on the original image
cv2.drawContours(image, contours, -1, (0, 255, 0), 2)
# Step 5: Display the image with contours using matplotlib
plt.imshow(image)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB for correct display
plt.axis() # Hide the axes
plt.show()
```

```python

import cv2
import matplotlib.pyplot as plt
# Step 1: Read the image and convert to grayscale

image = cv2.imread('sunflower.jpg') # Replace with your image path
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Step 2 : Apply thresholding
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
# Step 3: Find contours
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# Step 4: Draw contours on the original image
cv2.drawContours(image, contours, -1, (0, 255, 0), 2)
# Step 5: Display the image with contours using matplotlib
plt.imshow(image)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB for correct display
plt.axis() # Hide the axes
plt.show()
```
